```python
import os, json
import pandas as pd
import matplotlib.pyplot as plt

path_to_json = '../benchmark_results'
from glob import glob
results = []
for f_name in glob('../JSON/JSON_results/*.json'):
    with open('../benchmark_results/' + f_name) as f:
        results.append(json.load(f))
```

```python
results
```

```python
benchmarks = set(result['benchmark_name'] for result in results)
print(benchmarks)
```

```python
for name in benchmarks:
    entries = list(result for result in results if result['benchmark_name'] == name)
    print(name)
    frame = pd.DataFrame(data = entries)
    frame.plot(x = 'number_layers', y = 'amat')
    frame.plot(x = 'number_layers', y = 'amat_read')
```
